```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
plt.plot([1,2,3],[4,5,6]) # Plotting lines
plt.show()                # Displaying the chart
```

```python
plt.plot([1,2,3],[5,7,4]) # Plotting lines
plt.show()                # Displaying the chart
```

#### *Multiple plots*

```python
x1,y1 = [1,2,3],[5,7,4]
plt.plot(x1,y1,label = 'First Line')
x2,y2 = [1,2,3],[10,11,14]
plt.plot(x2,y2,label = 'Second Line')

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title('Graph')

plt.legend()        # Involves a legend

plt.show()          # Display the char
```

#### *Multiple views*

```python
t = np.arange(0.0,20.0,1)  # Creates a numeric sequence from 0.0 to 19.0 (20.0 is not included) ([start,]end[,step])
s1 = np.arange(1.0,21.0,1)
s2 = np.arange(4.0,24.0,1)

plt.subplot(2,1,1)
plt.plot(t,s1)
plt.xlabel('Item(s)')
plt.ylabel('Value')
plt.title('Chart')
plt.grid(True)

plt.subplot(2,1,2)
plt.plot(t,s2)
plt.xlabel('Item(s)')
plt.ylabel('Value')
plt.grid(True)

plt.show()
```

```python
def fnplot(l):
    plt.plot(l)
    plt.title('List')
    plt.xlabel('Index')
    plt.ylabel('Value')
    plt.show()

fnplot([50,50,50,65,65,75,75,80,80,90,90,90,90])
```

```python
#Plotting sine wave using line chart

x = np.arange(-10,10.01,0.01)
y = np.sin(x)
plt.plot(x,y)
plt.title("y = sin(x)")
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
x = np.arange(-math.pi,math.pi+0.01,0.01)
y = np.sin(x)
plt.plot(x,y,label = 'y = sin(x)')
y = np.cos(x)
plt.plot(x,y,label = 'y = cos(x)')

plt.title("Trigonometric Curves")
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
#Plotting Quadratic equation using line chart

x = np.arange(-10,10.01,0.01)
y = x**2
plt.plot(x,y)
plt.title("y = x²")
plt.xlabel('x')
plt.ylabel('y')
plt.show()

y = np.arange(-10,10.01,0.01)
x = y**2
plt.plot(x,y)
plt.title("y² = x")
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
y = np.arange(1,5)
plt.plot(y,'y')      # yellow line
plt.plot(y+1,'w')    # white line
plt.plot(y+2,'r')    # red line
plt.plot(y+3,'m')    # magenta line
plt.plot(y+4,'b:')   # black dotted line
plt.plot(y+5,'g-.')  # green dash-dot line
plt.plot(y+6,'c--')  # cyan dashed line
plt.plot(y+7,'b-')   # blue solid line
plt.show()
```

#### *Bar plot/chart*

```python
y_axis = [20,50,30]
x_axis = range(len(y_axis))
plt.bar(x_axis,y_axis,width=0.5,color="teal")
plt.show()
```

```python
objects = ('Object A','Object B','Object C','Object D','Object E')
x_pos = np.arange(len(objects))
values = [10,5,22,17,10]

plt.bar(x_pos, values, align='center',color='purple') 
plt.xticks(x_pos,objects)                                # Set location and label
plt.xlabel('Values')
plt.ylabel('Objects')

plt.show()
```

```python
# Plot a bar chart horizontally
objects = ('Object A','Object B','Object C','Object D')
y_pos = np.arange(len(objects))
values = [10,5,22,17]

plt.barh(y_pos, values, align='center',color='seagreen')
plt.yticks(y_pos,objects)
plt.xlabel('Values')
plt.ylabel('Objects')

plt.show()
```

```python
# Plotting two list elements using bar chart
x1,y1 = [2,4,6,8,10],[6,7,8,2,4]
x2,y2 = [1,3,5,7,9] ,[7,8,2,4,2]

plt.bar(x1,y1,label="Bars1",color="black")
plt.bar(x2,y2,label="Bars2",color="grey")
plt.xlabel('x')
plt.ylabel('y')
plt.title('Bar Graph\nWith Multiple Title')

plt.legend()
plt.show()
```

#### *Pie plot/chart*

```python
labels = ('Object A','Object B','Object C','Object D')
sizes = [215,123,334,567]
colors = ['gold','yellowgreen','lightcoral','lightskyblue']
explode = (0.145,0,0.1001,0)                                         # Explode first and third slice

plt.pie(sizes, explode=explode,labels=labels,colors=colors,shadow=True,startangle=140)

plt.axis('equal')
plt.show()
```

```python
labels = ('Object A','Object B','Object C','Object D')
sizes = [215,123,334,567]
colors = ['gold','yellowgreen','lightcoral','lightskyblue']
explode = (0,0,0.2,0)                                         # Explode third slice

plt.pie(sizes, explode=explode,labels=labels,colors=colors)

plt.axis('equal')
plt.show()
```
