```python
data = pd.read_csv('countries.csv')
data
```

### Compare the population growth in US and India

```python
data.country == 'United States'
```

```python
us = data[data.country == 'United States']
us
```

```python
ind = data[data.country == 'India']
ind
```

### Population Growth - India and US

```python
plt.plot(ind.year, ind.population/10**6, 'green')
plt.plot(us.year,  us.population/10**6,  'red'  )
plt.title('Population Growth - India and US')
plt.legend(['India','United States'])
plt.xlabel('Year')
plt.ylabel('Population (in millions)')
plt.show()
```

### Percentage Population Growth relative to Population of First Year - India and US

```python
plt.plot(ind.year, (ind.population/ind.population.iloc[0])*100, 'green')
plt.plot(ind.year, (us.population/us.population.iloc[0])*100,   'red'  )
plt.title('Percentage Population Growth relative to Population of First Year - India and US')
plt.legend(['India','United States'])
plt.xlabel('Year')
plt.ylabel('Percentage Population Growth')
plt.show()
```
